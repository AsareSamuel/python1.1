> ### Create a conditional column to record died for 0 field entries and survived for 1 field entries. Since its under the numpy library, import numpy

```python
import numpy as np
```

> ### Then write the code to make 0 died and 1 survived 

```python
np.where(df['survived']==0,'Died','Survived')
```

> ### That gives us the above, but the column is not quite created yet, so we create the column with the code below

```python
df['fate']=np.where(df['survived']==0,'Died','Survived')
```

> ### Replace values in the pclass column.

```python
df['pclass']=df['pclass'].replace(1,'first class').replace(2,'second class').replace(3,'third class')
```
